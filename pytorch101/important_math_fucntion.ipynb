```python
import numpy as np
import matplotlib.pyplot as plt
import torch
```

```python
a = torch.arange(0., 60.)
a
```

```python
a_cos = torch.acos(a)
```

```python
plt.plot(a, a_cos)
plt.show()
```

```python
deg = np.arange(0., 91., 1)
deg
```

```python
val = np.cos(deg)
val
```

```python
plt.plot(deg, val)
plt.axis("tight")
plt.show()
```

```python
a_cos = np.arccos(val)
a_cos
```

```python
plt.plot(val, a_cos)
plt.axis("tight")
plt.show()
```
